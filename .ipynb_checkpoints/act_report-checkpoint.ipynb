This Wrangle and Analyze Data Project is part of Udacity's Nanodegree Data Analyst program. This project involved wrangling (and analyzing and visualizing) data from the Twitter user @ WeRateDogs, collected from different sources connected with the tweets. WeRateDogs is a Twitter account that rates photos of people's dogs in a humorous way. These ratings almost always have a denominator of 10. 

Quality performance, and tidiness problems were carried out and evaluated and then cleaned after the data had been merged. Finally, different insights and visualizations were created, and can be found below.

Insight 6: Correlation between retweet and Favourite count
<br>

    

    
From the bar chat graph above, it clearly shown that the counts of favourites are more than the counts of retweet.

Because the most favorited and most retweeted tweets are the same, one can deduce from the secondgraph above that a higher number of retweets increases the likelihood that the tweet will also be favorited. Based on the best fit line in the graph, there is a sharp, positive connection. Every rise in the number of retweets will result in several favourite hits.

```python
# Importing all packages needed
import re
import json
import time
import requests
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
import tweepy as tweepy
import matplotlib.pyplot as plt
from tweepy import OAuthHandler
from tweepy.parsers import JSONParser
from timeit import default_timer as timer
```

```python
# Reading the twitter archive enhanced file
df_twitter_archive = pd.read_csv('/home/axe_rhitta/W_A/twitter-archive-enhanced.csv')
df_twitter_archive.sort_values('timestamp')
df_twitter_archive.head()
```

### 2. Use the Requests library to download the tweet image prediction (image_predictions.tsv)

```python
# Gathering image predictions file from Udacity's servers.
url = "https://d17h27t6h515a5.cloudfront.net/topher/2017/August/599fd2ad_image-predictions/image-predictions.tsv"
response = requests.get(url, allow_redirects=True)

with open('image-predictions.tsv', mode ='wb') as file:
    file.write(response.content)

#Read TSV file
df_image_prediction = pd.read_csv('image-predictions.tsv', sep='\t' )
df_image_prediction.head()
```

### 3. Use the Tweepy library to query additional data via the Twitter API (tweet_json.txt)

```python
# Accessing twitter archives using your unique keys and tokens
consumer_key = ''
consumer_secret = ''
access_token = ''
access_secret = ''

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

api = tweepy.API(auth, parser=JSONParser(), wait_on_rate_limit=True)
```

```python
# Testing running the API 
tweet = api.get_status(887473957103951883,tweet_mode='extended')
print(tweet['id'])
print(tweet['retweet_count'])
```

```python
# Querying the API for JSON data for each tweet ID in the Twitter archive and the Track timing of tweet completion as well.
df_list = []
errors_list = []

tweets = df_twitter_archive['tweet_id']

now = time.time()

for index, t in enumerate(tweets):

    try:
        start = time.time()
        tweet = api.get_status(t,tweet_mode='extended')
        end = time.time()
        print("Reading Tweet {}/{}. ({}%) complete. Elapsed Time = {}".format((index+1),len(tweets), round(((index+1)/len(tweets)) * 100,1), round((end - start),2)))
        retweet_count = tweet['retweet_count']
        favorite_count = tweet['favorite_count']
        df_list.append({'tweet_id':int(t),
                        'retweet_count':int(retweet_count),
                        'favorite_count':int(favorite_count)
                       })

    
    
    except Exception as e:
        print(str(t) + str(e))
        errors_list.append(t)

then = time.time()
diff = round(then - now)
minutes = diff // 60
seconds = diff % 60
print("Reading Process Completed. Elapsed Time is {}:{}".format(minutes, seconds))
```

```python
errors_list
```

```python
# Saving the df of queried tweets to JSON and then convert to a dataframe
df_json = pd.DataFrame(df_list, columns = ['tweet_id', 'retweet_count', 'favorite_count'])
df_json.to_json('tweet_json.txt')
df_json = pd.read_json('tweet_json.txt')
df_json.head()                                                         
```

#### Define: Creating one column for the various dog types (doggo, floofer, pupper, puppo).

#### Code

```python
# First, we will substitute all the NaN for null values. 
dog_twitterpage['doggo'].replace(np.nan,'', inplace=True)
dog_twitterpage['floofer'].replace(np.nan,'', inplace=True)
dog_twitterpage['pupper'].replace(np.nan,'', inplace=True)
dog_twitterpage['puppo'].replace(np.nan, '', inplace=True)
```

```python
dog_twitterpage['dog_type'] = 'None'

def get_dog_stage(row):
    dog_type = []
    if row['doggo'] == 'doggo':
        dog_type.append('doggo')
    if row['floofer'] == 'floofer':
        dog_type.append('floofer')
    if row['pupper'] == 'pupper':
        dog_type.append('pupper')
    if row['puppo'] == 'puppo':
        dog_type.append('puppo')
    
    if len(dog_type) < 1:
        row['dog_type'] = 'None'
    else:
        row['dog_type'] = ','.join(dog_type)
        
    return row

dog_twitterpage = dog_twitterpage.apply(get_dog_stage, axis=1)
```

```python
dog_twitterpage.dog_type.value_counts()
```

```python
dog_twitterpage['dog_type'].replace('doggo,pupper','doggo-pupper', inplace=True)
dog_twitterpage['dog_type'].replace('doggo,floofer','doggo-floofer', inplace=True)
dog_twitterpage['dog_type'].replace('doggo,puppo','doggo-puppo', inplace=True)
```

#### Visualization 2: Correlation between retweet and Favourite count. 

```python
plt.figure(figsize = [10,6])
width = 0.4
bins = np.arange(dog_twitterpage['favorite_count'].min() , dog_twitterpage['favorite_count'].max() + 500, 500)
dog_twitterpage.favorite_count.plot(kind='hist',  bins=bins )
bins = np.arange(dog_twitterpage['retweet_count'].min() , dog_twitterpage['retweet_count'].max() + 500, 500)
dog_twitterpage.retweet_count.plot(kind='hist', bins=bins )

plt.xlabel('count')
plt.legend()
plt.xlim(200,30000)
plt.ylim(0,300)
plt.title( 'favorite/retweet');
```

```python
# Make a new dataframe with only the columns we need
CRF = dog_twitterpage[['retweet_count', 'favorite_count']]

# We will use seaborn to make a scatterplot with a regression line
sns.regplot(CRF['retweet_count'], CRF['favorite_count']).set_title(
    "Correlation Between Retweet and Favorite Count", size = 15)

# Save the plot so that we can use it for the separate reports
plt.savefig('Correlation Between Retweet and Favorite Count.jpg')
```

### Observation
From the bar chat graph above, it clearly shown that the counts of favourites are more than the counts of retweet.

Because the most favorited and most retweeted tweets are the same, one can deduce from the secondgraph above that a higher number of retweets increases the likelihood that the tweet will also be favorited. Based on the best fit line in the graph, there is a sharp, positive connection. Every rise in the number of retweets will result in several favourite hits.
